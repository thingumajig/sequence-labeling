```python
def dump_meta(g, tag='ner'):
    for p in g:
        if type(p)==Group:
            dump_meta(p.metas, tag)
        else:
            if tag in p.tags:
                print(f'''
                title: {p.title}
                url: {p.url}
                description: {p.description}
                stats: {p.stats}
                tags: {p.tags}
                ''')
                
dump_meta(METAS, tag = 'ner')
```
